```python
import pandas as pd  # convention to import and use pandas like this
import numpy as np #numpy import

df = pd.read_csv("hgdp-ceph-geno.out",sep="\t") #importing hgdp genotype dataset
```

```python
pdf = pd.read_csv("hgdp-ceph-unrelated.out",sep="\t") #reading hgdp file which has list of population
```

```python
distdata = pdf[pdf["Geographic origin"].isin(["Central African Republic","Nigeria","Democratic Republic of Congo", "Kenya","Namibia","Senegal","South Africa","China","Pakistan","Cambodia","Japan","Siberia","France","Italy","Orkney Islands","Italy (Bergamo)","Russia","Russia Caucasus"])]
np.unique(distdata.population)
```

```python
#closedata = pdf[pdf["Region"].isin(["Europe","Middle Est"])]
clodata = pdf[pdf["Geographic origin"].isin(["France","Italy","Italy (Bergamo)","Orkney Islands","Russia", "Russia Caucasus", "Israel (Carmel)", "Israel (Negev)"])]
np.unique(clodata.population)
```

```python
distantdata = df.merge(distdata, on='hgdp_id',how='inner')
#To keep only rows that match from the data frames, specify the argument how=‘inner’.
closedata = df.merge(clodata, on='hgdp_id',how='inner')
#To keep only rows that match from the data frames, specify the argument how=‘inner’.
```

```python
closedata
```

```python
#tracking of snp data
snpdf = pd.read_csv("hgdp-ceph-marker.out",sep="\t") 
#marker only with snp is used ie, 0 or 1 or 2
snpdf = snpdf[snpdf.type_marker == 'snp']
used_snp = pd.DataFrame(snpdf, columns = ["mkr_ceph_id"])
```

```python
#bringing snp id to our dataset
distantdata = used_snp.merge(distantdata, on = "mkr_ceph_id", how = "inner")
closedata = used_snp.merge(closedata, on = "mkr_ceph_id", how = "inner")
```

```python
distantdata
```

```python
distantdata['marker'] = [2 if(x ==1 and y==1) else 1 if((x ==1 and y== 2)or(x ==2 and y == 1)) else 0 for x,y in zip(distantdata['allele1'],distantdata['allele2'])]
closedata['marker'] = [2 if(x ==1 and y==1) else 1 if((x ==1 and y== 2)or(x ==2 and y == 1)) else 0 for x,y in zip(closedata['allele1'],closedata['allele2'])]
#observing allele1 and allele2, of 1 and 1 ==> 2 if 1 and 2 or 2 and 1 ==> 1 otherwise 0 (ie, last case if 2 and 2)
```

```python
distantdata #observe the marker column here
```

```python
#converting data into ancestrymap genotype as described in eigenstart
distance_ancestry = pd.DataFrame(distantdata, columns = ["mkr_ceph_id", "hgdp_id","marker"])
close_ancestry = pd.DataFrame(closedata, columns = ["mkr_ceph_id", "hgdp_id","marker"])
```
