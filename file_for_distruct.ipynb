```python
import pandas as pd
```

```python
df = pd.read_csv("k_mean_test.txt",sep="\t",) 
```

```python
df = df.rename(columns={"X.eigvals.":"HGDP"})
df
```

```python
import numpy as np
origin = pd.unique(df.Origin)
```

```python
df[df['cluster'] == 3]
```

```python
df['Cluster1'] = np.where(df['cluster']==1, 1, 0)
df['Cluster2'] = np.where(df['cluster']==2, 1, 0)
df['Cluster3'] = np.where(df['cluster']==3, 1, 0)
df
```

```python
df_french = df.query('Origin == "French"', inplace=False)
```

```python
av_column = df_french.mean(axis=0)
```

```python
newdf = df.groupby(['Origin'])['Cluster1','Cluster2','Cluster3'].mean()
```

```python
newdf
```

```python
countdf = df.groupby(['Origin'])['Origin'].count()
newdf = newdf.join(countdf)
```

```python
newdf = newdf.rename(columns={"Origin":"Average"})
newdf
```

```python
newdf.insert(0, 'NewID', range(1, 1 + len(newdf)))
newdf
```

```python
newdf['NewID'] =newdf['NewID'].astype(str)+':'
```

```python
newdf
```

```python
newdf = newdf.style.format({
    'Cluster1': '{:,.2f}'.format,
    'Cluster2': '{:,.2f}'.format,
    'Cluster3': '{:,.2f}'.format,
})
```
