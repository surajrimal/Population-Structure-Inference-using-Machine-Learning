```python
import pandas as pd
```

```python
df = pd.read_csv("soft_kmeans_mclust_2_clus_result.txt",sep="\t",) 
#df = df.rename(columns={"X.eigvals.":"HGDP"})
#df
```

```python
import numpy as np
```

```python
sdf = pd.read_csv("sorted.txt",sep="\t",names=['test'], header = None) 
#sdf
sorter = sdf.test.tolist()
df.Origin = df.Origin.astype("category")
df.Origin.cat.set_categories(sorter, inplace=True)
df = df.sort_values(["Origin"]) 
df
```

```python
df['Cluster1'] = np.where(df['cluster']==1, 1, 0)
df['Cluster2'] = np.where(df['cluster']==2, 1, 0)
#df
```

```python
names_file = pd.read_csv("casia_distant.names",sep="\t",header = None) 
#names_file
```

```python

names_file.columns = ['ID', 'Origin']
names_file
```

```python
named_file = pd.merge(df, names_file, on='Origin', how='left')
named_file
```

```python
named_file['ID'] =named_file['ID'].astype(str)+' :'
named_file.insert(0, 'SN', range(1, 1 + len(named_file)))
named_file['Miss'] = "(0)"

#making casia.indvq like file
```

```python
named_file
```

```python
final_file = pd.DataFrame(named_file, columns = ["SN","hgdp_id","Miss","ID","Cluster1","Cluster2","Cluster3"])
final_file
```

```python
final_file.to_csv ('casia_distant.indivq', index = None, header=False, sep = '\t')
```
