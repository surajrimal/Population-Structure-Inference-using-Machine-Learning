```python
import pandas as pd
```

```python
df = pd.read_csv("soft_kmeans_mclust_3_clus_result.txt",sep="\t",) 
```

```python
df = df.rename(columns={"Geographic.origin":"Country"})
df
```

```python
import numpy as np
origin = pd.unique(df.Origin)
```

```python
sdf = pd.read_csv("sorted.txt",sep="\t",names=['test'], header = None) 
sdf
sorter = sdf.test.tolist()
#sorter
```

```python
#sorter
df.Origin = df.Origin.astype("category")
df.Origin.cat.set_categories(sorter, inplace=True)
df = df.sort_values(["Origin"]) 
df
```

```python
df['Cluster1'] = np.where(df['cluster']==1, 1, 0)
df['Cluster2'] = np.where(df['cluster']==2, 1, 0)
df['Cluster3'] = np.where(df['cluster']==3, 1, 0)
df
```

```python
newdf = df.groupby(['Origin'],sort = False)['Cluster1','Cluster2','Cluster3'].mean()
```

```python
newdf
```

```python
countdf = df.groupby(['Origin'],sort = False)['Origin'].count()
newdf = newdf.join(countdf)
```

```python
newdf = newdf.rename(columns={"Origin":"Average"})
newdf
```

```python
newdf.insert(0, 'NewID', range(1, 1 + len(newdf)))
```

```python
names_file = pd.DataFrame(newdf, columns = ['NewID'])
names_file
```

```python
newdf['NewID'] =newdf['NewID'].astype(str)+':'
```
